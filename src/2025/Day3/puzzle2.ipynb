```python
# Read file 'input_day1' and return a list with each line as an element
with open("resources/input_day3.txt", "r", encoding="utf-8") as f:
    lines = f.read().splitlines()
```

```python
def get_bank_highest_joltage(bank: list[str], nb_digit: int) -> str:
    """Get the highest joltage from the bank of batteries."""
    # If looking for only one number, return the max.
    if nb_digit == 1:
        return max(bank)
    
    # We avoid the n last element since it's a n number combination and need the last
    # digits to make the joltage (batteries can't be re-arranged).
    first_number = max(bank[:-(nb_digit - 1)])
    # The second is then the highest number after the first one.
    # The "index" method for lists, returns the first index of the element found.
    return first_number + get_bank_highest_joltage(
            bank[bank.index(first_number)+1:], nb_digit - 1,
        )
```

```python
solution = 0
for b in lines:
    solution += int(get_bank_highest_joltage(b, 12))
print(f"Solution: {solution}")
```
