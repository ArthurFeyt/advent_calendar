```python
# Read file 'input_day1' and return a list with each line as an element
with open("resources/input_day3.txt", "r", encoding="utf-8") as f:
    lines = f.read().splitlines()
```

```python
def get_bank_highest_joltage(bank: list[str]) -> int:
    """Get the highest joltage from the bank of batteries."""
    # We avoid the last element since it's a 2 number combination.
    first_number = max(bank[:-1])
    # The second is then the highest number after the first one.
    # The "index" method for lists, returns the first index of the element found.
    second_number = max(bank[bank.index(first_number)+1:])
    return int(first_number + second_number)
```

```python
solution = 0
for b in lines:
    solution += get_bank_highest_joltage(b)
print(f"Solution: {solution}")
```
