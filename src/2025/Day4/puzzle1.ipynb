```python
# Read file 'input_day1' and return a list with each line as an element
with open("resources/input_day4.txt", "r", encoding="utf-8") as f:
    lines = f.read().splitlines()

lines[0][2]
```

```python
from itertools import product

def get_paper_rolls(matrix: list[str]) -> int:
    m = len(matrix)
    n = len(matrix[0])
    total = 0
    for i in range(m):
        for j, e in enumerate(matrix[i]):
            if e == ".":
                continue
            not_paper_rolls = 0
            i_positions = [0]
            j_positions = [0]

            if i==0:
                not_paper_rolls += 3
                i_positions.append(1)
            elif i==m-1:
                not_paper_rolls += 3
                i_positions.append(-1)
            else:
                i_positions.append(-1)
                i_positions.append(1)

            if j==0:
                not_paper_rolls += 3
                j_positions.append(1)
            elif j==n-1:
                not_paper_rolls += 3
                j_positions.append(-1)
            else:
                j_positions.append(-1)
                j_positions.append(1)

            for a, b in product(i_positions, j_positions):
                if matrix[i+a][j+b] == ".":
                    not_paper_rolls += 1
                if not_paper_rolls > 4:
                    total += 1
                    break
    return total
            
```

```python
get_paper_rolls(lines)
```
