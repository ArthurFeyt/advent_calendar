```python
# Read file 'input_day2' and return a list with each line as an element
with open("resources/input_day2.txt", "r", encoding="utf-8") as f:
    ranges = f.read().split(",")

ranges = [r.split("-") for r in ranges]
ranges = [(int(r[0]), int(r[1])) for r in ranges]
```

```python
def get_invalid_ids(range_id: tuple[int, int]) -> list[int]:
    invalid_ids = []
    
    for id in range(range_id[0], range_id[1] + 1):
        str_id = str(id)
        n = len(str_id)
       
        if n % 2 == 1:
            continue
       
        if str_id[:n//2] == str_id[n//2:]:
            invalid_ids.append(id)

    return invalid_ids
```

```python
test_ranges = [
    (11, 22), (95, 115), (998, 1012), (1188511880, 1188511890), (222220, 222224),
    (1698522, 1698528), (446443, 446449), (38593856, 38593862), (565653, 565659),
    (824824821, 824824827), (2121212118, 2121212124),
]

invalid_ids = []
for r_id in test_ranges:
    invalid_ids += get_invalid_ids(r_id)

print(invalid_ids)
print(f"Solution: {sum(invalid_ids)}")
```
